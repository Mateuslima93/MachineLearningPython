```python
import pandas as pd
import numpy as np
```

```python
uri = 'https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv'
df = pd.read_csv(uri)
df.head()
```

```python
df.rename(columns={'mileage_per_year':'quilometros_por_ano', 'model_year':'idade', 'price':'preco', 'sold':'vendido'},inplace=True)
df.head()
```

```python
from datetime import date
ano_atual = date.today().year
print(ano_atual)
df.vendido = df.vendido.apply(lambda x: 1 if x =='yes' else 0)
df.idade = df.idade.apply(lambda x: ano_atual - x)
df.quilometros_por_ano = df.quilometros_por_ano.apply(lambda x: int(round((x/1.60934))))
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.preprocessing import StandardScaler

x = df[['quilometros_por_ano','idade','preco']]
y = df['vendido']

seed = 101
np.random.seed(seed)
raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x,y,test_size=0.25, stratify=y)

escala = StandardScaler()
escala.fit(raw_treino_x)
treino_x = escala.transform(raw_treino_x)
teste_x = escala.transform(raw_teste_x)

modelo = SVC()
modelo.fit(treino_x,treino_y)
previsao = modelo.predict(teste_x)
acuracia = accuracy_score(teste_y,previsao)*100

print(f'A acurácia do modelo é de {acuracia.round(3)%}')
```

```python
dummy = DummyClassifier()

dummy.fit(treino_x,treino_y)
previsao_dummy = dummy.predict(teste_x)
acuracia_dummy = accuracy_score(teste_y,previsao_dummy)*100

print(f'A acurácia do modelo dummy é de {acuracia_dummy.round(3)}%')
```
